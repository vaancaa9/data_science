```python
import streamlit as st
import mysql.connector

# Function to fetch data from MySQL tables based on the selected question
def get_result(question):
    try:
        # Connect to MySQL database
        conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="test",
            database="retail_orders"
        )
        cursor = conn.cursor(dictionary=True)

        # Define the query based on the selected question
        query = None

        if question == "Find all orders with their corresponding shipping mode and region?":
            query = """
                SELECT c.order_id, c.ship_mode, c.region
                FROM customer_order AS c
            """

        elif question == "List all products with their cost price, sale price, and profit for each order?":
            query = """
                SELECT co.order_id, co.product_id, co.cost_price, co.sale_price, co.profit
                FROM customers_order AS co
            """

        elif question == "Find the total profit generated for each region?":
            query = """
                SELECT c.region, SUM(co.profit) AS total_profit
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.region
            """

        elif question == "Identify the top 3 regions with the highest number of orders?":
            query = """
                SELECT c.region, COUNT(c.order_id) AS order_count
                FROM customer_order AS c
                GROUP BY c.region
                ORDER BY order_count DESC
                LIMIT 3
            """

        elif question == "Calculate the total revenue generated for each category?":
            query = """
                SELECT c.category, SUM(co.sale_price) AS total_revenue
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.category
            """

        elif question == "Find the top 5 products with the highest profit margin (profit/cost price)?":
            query = """
                SELECT co.product_id, (co.profit / co.cost_price) AS profit_margin
                FROM customers_order AS co
                WHERE co.cost_price > 0
                ORDER BY profit_margin DESC
                LIMIT 5
            """

        elif question == "Determine the average discount percentage given for each sub-category?":
            query = """
                SELECT c.sub_category, AVG(co.discount_percent) AS avg_discount
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.sub_category
            """

        elif question == "Find all orders shipped via a specific mode (e.g., 'Express Shipping') and their corresponding profits?":
            query = """
                SELECT c.order_id, c.ship_mode, co.profit
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                WHERE c.ship_mode = 'Express Shipping'
            """

        elif question == "Identify the segment with the highest total revenue?":
            query = """
                SELECT c.segment, SUM(co.sale_price) AS total_revenue
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.segment
                ORDER BY total_revenue DESC
                LIMIT 1
            """

        elif question == "List all products sold in a specific region (e.g., 'West') along with their sale prices and discounts?":
            query = """
                SELECT co.product_id, co.sale_price, co.discount_amount
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                WHERE c.region = 'West'
            """

        elif question == "Find top 10 highest revenue generating products?":
            query = """
                SELECT co.product_id, SUM(co.sale_price) AS total_revenue
                FROM customers_order AS co
                GROUP BY co.product_id
                ORDER BY total_revenue DESC
                LIMIT 10
            """

        elif question == "Find the top 5 cities with the highest profit margins?":
            query = """
                SELECT c.city, SUM(co.profit) AS total_profit
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.city
                ORDER BY total_profit DESC
                LIMIT 5
            """

        elif question == "Calculate the total discount given for each category?":
            query = """
                SELECT c.category, SUM(co.discount_amount) AS total_discount
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.category
            """

        elif question == "Find the average sale price per product category?":
            query = """
                SELECT c.category, AVG(co.sale_price) AS avg_sale_price
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.category
            """

        elif question == "Find the region with the highest average sale price?":
            query = """
                SELECT c.region, AVG(co.sale_price) AS avg_sale_price
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.region
                ORDER BY avg_sale_price DESC
                LIMIT 1
            """

        elif question == "Find the total profit per category?":
            query = """
                SELECT c.category, SUM(co.profit) AS total_profit
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.category
            """

        elif question == "Identify the top 3 segments with the highest quantity of orders?":
            query = """
                SELECT c.segment, COUNT(c.order_id) AS order_count
                FROM customer_order AS c
                GROUP BY c.segment
                ORDER BY order_count DESC
                LIMIT 3
            """

        elif question == "Determine the average discount percentage given per region?":
            query = """
                SELECT c.region, AVG(co.discount_percent) AS avg_discount
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.region
            """

        elif question == "Find the product category with the highest total profit?":
            query = """
                SELECT c.category, SUM(co.profit) AS total_profit
                FROM customer_order AS c
                JOIN customers_order AS co ON c.order_id = co.order_id
                GROUP BY c.category
                ORDER BY total_profit DESC
                LIMIT 1
            """

        elif question == "Calculate the total revenue generated per year?":
            query = """
                SELECT YEAR(co.order_date) AS year, SUM(co.sale_price) AS total_revenue
                FROM customers_order AS co
                GROUP BY year
                ORDER BY year
            """

        else:
            st.error("Query for this question is not defined yet.")
            return None

        # Execute the query
        cursor.execute(query)
        result = cursor.fetchall()
        conn.close()

        return result

    except mysql.connector.Error as e:
        st.error(f"Error: {e}")
        return None

# Streamlit App
def main():
    st.title("Question & Answer App")

    # List of 20 questions
    questions = [
        "Find all orders with their corresponding shipping mode and region?",
        "List all products with their cost price, sale price, and profit for each order?",
        "Find the total profit generated for each region?",
        "Identify the top 3 regions with the highest number of orders?",
        "Calculate the total revenue generated for each category?",
        "Find the top 5 products with the highest profit margin (profit/cost price)?",
        "Determine the average discount percentage given for each sub-category?",
        "Find all orders shipped via a specific mode (e.g., 'Express Shipping') and their corresponding profits?",
        "Identify the segment with the highest total revenue?",
        "List all products sold in a specific region (e.g., 'West') along with their sale prices and discounts?",
        "Find top 10 highest revenue generating products?",
        "Find the top 5 cities with the highest profit margins?",
        "Calculate the total discount given for each category?",
        "Find the average sale price per product category?",
        "Find the region with the highest average sale price?",
        "Find the total profit per category?",
        "Identify the top 3 segments with the highest quantity?",
        "Determine the average discount percentage given per region?",
        "Find the product category with the highest total profit?",
        "Calculate the total revenue generated per year?"
        ]


    # Dropdown for selecting a question
    selected_question = st.selectbox("Enter your question:", questions)

    # Display result based on selected question
    if st.button("Get Result"):
        result = get_result(selected_question)
        if result:
            st.table(result)
        else:
            st.write("No results found for the selected question.")

if __name__ == "__main__":
    main()

import streamlit as st
import pandas as pd
from sqlalchemy import create_engine

# Function to create a database connection
def create_connection():
    try:
        # Update these credentials as per your database
        engine = create_engine(
            "mysql+pymysql://root:test@localhost:3306/retail_orders"
        )
        return engine
    except Exception as e:
        st.error(f"Error connecting to the database: {e}")
        return None

# Define the queries
def get_total_sales_by_region(engine):
    query = """
        SELECT c.region, SUM(co.sale_price) AS total_sales
        FROM customer_order AS c
        JOIN customers_order AS co ON c.order_id = co.order_id
        GROUP BY c.region;
    """
    return pd.read_sql(query, engine)

def get_most_profitable_category(engine):
    query = """
        SELECT c.category, SUM(co.profit) AS total_profit
        FROM customer_order AS c
        JOIN customers_order AS co ON c.order_id = co.order_id
        GROUP BY c.category
        ORDER BY total_profit DESC
        LIMIT 1;
    """
    return pd.read_sql(query, engine)

def get_average_discount(engine):
    query = """
        SELECT AVG(co.discount_percent) AS average_discount
        FROM customers_order AS co;
    """
    return pd.read_sql(query, engine)

# Streamlit App
st.title("Retail Orders Dashboard")
st.write("This app connects to the `retail_orders` database and answers specific business questions.")

# Connect to the database
engine = create_connection()

if engine:
    # Question 1: Total Sales by Region
    st.subheader("1. Total Sales by Region")
    total_sales = get_total_sales_by_region(engine)
    st.dataframe(total_sales)

    # Question 2: Most Profitable Product Category
    st.subheader("2. Most Profitable Product Category")
    most_profitable = get_most_profitable_category(engine)
    st.dataframe(most_profitable)

    # Question 3: Average Discount Applied
    st.subheader("3. Average Discount Applied Across All Orders")
    average_discount = get_average_discount(engine)
    st.write(f"Average Discount: {average_discount['average_discount'].iloc[0]:.2f}%")
else:
    st.error("Failed to connect to the database.")

```
